```python
'''
A : 100 - 80
B : 79 - 70
C : 69 - 60
D : 59 - 50
E : 49 - 0
'''

score = int(input("Enter your score : "))
if score >= 80:
    print("Grade A")
elif score >= 70:
    print("Grade B")
elif score >= 60:
    print("Grade C")
elif score >= 50:
    print("Grade D")
else:
    print("Grade E")
```
